## To reduce the number of intensity levels in an image from 256 to 2, in integer powers of 2. 

```python
#To reduce the number of intensity levels in an image from 256 to 2, in integer powers of 2. 

image = data.camera()
levels=int(input("Enter: "))
#reducing the number of distinct colors in an image, also known as color quantization
dev =255/(levels-1)
newIm=np.round(image/dev)*dev
fig,arr = plt.subplots(1,2)
arr[0].imshow(image)
arr[1].imshow(newIm)
```

# applied is a 2D convolution filter with a kernel 

```python
image = data.camera()
kernel2 = np.array([[-1,-1,-1],
                    [-1,8,-1],
                    [-1,-1,-1]])

f,arr = plt.subplots(1,4)
image3K = cv2.blur(image,(3,3))
image10k = cv2.blur(image,(10,10))
# applied is a 2D convolution filter with a kernel 
image20k = cv2.filter2D(src=image,ddepth=-1,kernel=kernel2)

plt.gray()

arr[0].imshow(image)
arr[0].set_title('original')


arr[1].imshow(image3K)
arr[1].set_title('3x3 kernal')

arr[2].imshow(image10k)
arr[2].set_title('10x10 kernal')

arr[3].imshow(image20k)
arr[3].set_title('20x20 kernal')
```

# Rotate an image by 45 and 90 degrees.

```python
img = data.camera()

print(f"Image shape : {img.shape}")
print(f"Image datatype : {type(img)}")

plt.gray()
f,arr = plt.subplots(1,3)
arr[0].imshow(img)

# Get the height and width of the image
height, width = img.shape[:2]

# Specify the rotation angle
angle_45 = 45
angle_90 = 90

# Specify the center of the image
center = (width/2, height/2)

# Perform the rotation
rotation_matrix_45 = cv2.getRotationMatrix2D(center, angle_45, 1.0)
rotation_matrix_90 = cv2.getRotationMatrix2D(center, angle_90, 1.0)
rotated_img_45 = cv2.warpAffine(img, rotation_matrix_45, (width, height))
rotated_img_90 = cv2.warpAffine(img, rotation_matrix_90, (width, height))

# Show the rotated image
arr[1].imshow(rotated_img_45)
arr[2].imshow(rotated_img_90)
cv2.waitKey(0) # waits for a key event
cv2.destroyAllWindows() # closes all the windows
```
