## 1. To reduce the number of intensity levels in an image from 256 to 2, in integer powers of 2. The desired number of intensity levels needs to be a variable input to your program. 

```python
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage import data
import numpy as np
import cv2

image = data.cat()
image=rgb2gray(image)
image =np.rint(image*256).astype(np.uint8)
plt.gray()
levels=2 
#reducing the number of distinct colors in an image, also known as color quantization
dev =255/(levels-1)
newIm=np.round(image/dev)*dev
fig,arr = plt.subplots(1,2,figsize=(10,9))
arr[0].imshow(image)
arr[0].set_title('Original Image')
arr[1].imshow(newIm)
arr[1].set_title('Converted Image')
```

# 3. Rotate an image by 45 and 90 degrees.

```python

fig,arr = plt.subplots(1,3,figsize=(20,10))
arr[0].imshow(image)
arr[0].set_title('Original Image')
# Get the height and width of the image
height, width = image.shape[:2]

# Specify the rotation angle
angle_45 = 45
angle_90 = 90

# Specify the center of the image
center = (width/2, height/2)

# Perform the rotation
rotation_matrix_45 = cv2.getRotationMatrix2D(center, angle_45, 1.0)
rotation_matrix_90 = cv2.getRotationMatrix2D(center, angle_90, 1.0)
rotated_img_45 = cv2.warpAffine(image, rotation_matrix_45, (width, height))
rotated_img_90 = cv2.warpAffine(image, rotation_matrix_90, (width, height))

# Show the rotated image
arr[1].imshow(rotated_img_45)
arr[1].set_title('45 deg rotated ')
arr[2].imshow(rotated_img_90)
arr[2].set_title('90 deg rotated ')

cv2.waitKey(0) # waits for a key event
cv2.destroyAllWindows() # closes all the windows
```
